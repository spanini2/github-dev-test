```python
import matplotlib.pyplot as plt
import numpy as np
import pandas
```

```python
x,y = np.meshgrid(np.linspace(-25,25,20),np.linspace(-25,25,20))

u = -y/np.sqrt(x**2 + y**2)
v = x/np.sqrt(x**2 + y**2)

n = -2
 
color = np.sqrt(((v-n)/2)*2 + ((u-n)/2)*2)

plt.quiver(x,y,u,v, color, alpha=0.8)
plt.show()
```
